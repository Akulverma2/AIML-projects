```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 🔹 Load dataset
df = pd.read_csv("spotify.csv")
# 🔹 Features
features = ["danceability", "energy", "valence", "tempo"]
# Remove missing values
df = df.dropna(subset=features)
```

```python
# 🔹 Scale features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])
# 🔹 Mood definitions
mood_dict = {
    "Happy": [0.8, 0.8, 0.9, 120],
    "Sad": [0.3, 0.3, 0.2, 70],
    "Energetic": [0.7, 0.9, 0.6, 140],
    "Calm": [0.4, 0.3, 0.5, 80]
}
```

```python
# 🔹 User input
user_mood = input("Enter mood (Happy/Sad/Energetic/Calm): ").capitalize()

if user_mood not in mood_dict:
    print("❌ Invalid mood!")
    exit()
# 🔹 Convert mood → scaled vector
mood_vector = pd.DataFrame(
    [mood_dict[user_mood]],
    columns=features
)

mood_vector = scaler.transform(mood_vector)
# 🔹 Compute similarity
similarity = cosine_similarity(mood_vector, scaled_features)
```

```python
# 🔹 Get top matches
top_indices = similarity[0].argsort()[-20:][::-1]
selected = np.random.choice(top_indices, 5, replace=False)
# 🔹 Display results
print(f"\n🎧 Top {user_mood} Songs For You:\n")
for i in selected:
    print(f"{df.iloc[i]['track_name']} - {df.iloc[i]['artist']}")
```
