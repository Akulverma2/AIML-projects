```python
import pandas as pd

# Import dataset
data = pd.read_csv("house_loan_checklist.csv")

# Basic information
print("Rows & Columns:", data.shape)
print("First 5 rows:\n", data.head())
print("Missing values:\n", data.isnull().sum())

# Numerical statistics
print("Sum:\n", data.sum(numeric_only=True))
print("Average:\n", data.mean(numeric_only=True))
print("Min:\n", data.min(numeric_only=True))
print("Max:\n", data.max(numeric_only=True))

# Export dataset
data.to_csv("exported_data.csv", index=False)
```
