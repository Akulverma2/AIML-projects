```python
from reportlab.pdfgen import canvas
from reportlab.lib.colors import red, blue, green, black
from PyPDF2 import PdfReader, PdfWriter
```

```python
# Create watermark PDF
def create_watermark(text, filename, color):
    c = canvas.Canvas(filename)

    c.setFont("Helvetica-Bold", 50)
    c.setFillColor(color)

    # Rotate for diagonal watermark
    c.translate(300, 400)
    c.rotate(45)

    c.drawCentredString(0, 0, text)
    c.save()
```

```python
# Apply watermark to your PDF
def add_watermark(input_pdf, watermark_pdf, output_pdf):
    reader = PdfReader(input_pdf)
    writer = PdfWriter()
    watermark = PdfReader(watermark_pdf).pages[0]

    for page in reader.pages:
        page.merge_page(watermark)
        writer.add_page(page)

    with open(output_pdf, "wb") as f:
        writer.write(f)
```

```python
# Customize here
create_watermark("NOT PUBLISHED", "watermark.pdf", red)

# Use your uploaded file name
add_watermark("Aiml_Assignment_2.pdf", "watermark.pdf", "final_output.pdf")

print("Done! Check final_output.pdf")
```
